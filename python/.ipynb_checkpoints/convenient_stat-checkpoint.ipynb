```python
%matplotlib inline
import pandas as pd
import matplotlib as plt
from pandas import DataFrame as df
```

```python
##1. csv 파일 불러오기 (convenient_store.csv)
con = pd.read_csv('convenient_store.csv')
con.head(5)
```

```python
##2. 전체 컬럼 정보, null 값 유무 확인
con.info()
```

```python
##3. 개수, 평균, 편차, 최소, 최대값 확인
# print(f'전체 데이터 개수: {len(con)}')
# print('시급 평균:', con['hourly_wage'].mean())
# print('시급 편차: ', con['hourly_wage'].std())
# print('최소 시급: ', con['hourly_wage'].min())
# print('최대 시급: ', con['hourly_wage'].max())

# int 64 형만 출력
con.describe()
```

```python
##4. 지역에 대한 통계, 개수, 유니크한 정보, 제일 빈도가 높은 지역
# print('지역 통계:\n', con['area1'].describe())
# print('지역 개수:\n',con['area1'].value_counts())
# print('유니크한 개수:',con['area1'].nunique())
# print('빈도가 제일 높은 지역: ',con['area1'].value_counts().head(1))
con.area.describe()
```

```python
##5. 시간 당 급여가 6500원 이상인 지역의 편의점 정보 출력 (상위 10개만)
# con[con['hourly_wage'] >= 6500].head(10)

high_wage = con[con.hourly_wage >= 6500]
high_wage.head(10)
```

```python
##6. 시간 당 급여가 높은 순서로 정렬 (sort_value() 함수 사용, 상위 10개만 출력)
# con.sort_values(by=['hourly_wage'], axis=0, ascending=False).head(10)

high_wage.sort_values(by='hourly_wage', ascending=False).head(10)
```

```python
##7. 영등포구에서 시간 당 급여가 6000원 이상인 편의점 검색
# con[con['hourly_wage'] >= 6000]['company']

y_high_wage = con[(con.hourly_wage >= 6000) & (con.area1 == '영등포구')]
y_high_wage
```

```python
##8. CU 편의점만 출력 (상위 10개만)
# con[con['name'] == 'CU'].head(10)

# obj -> str converting
cu = con[con.company.str.contains('CU')]
cu.head(10)
```

```python
##9. 지역 컬럼(location)을 추가한 다음, in Seoul 이라는 값 저장, 상위 5개 출력
con['location'] = 'in Seoul'
con.head(5)
```

```python
##10. 6000원 이상 컬럼 추가(more_than_6000) -> True, False 값 저장 (상위 20개 출력)
# con['more_than_6000'] = (con['hourly_wage'] >= 6000)
# con.head(20)

con['more_than_6000'] = con.hourly_wage >= 6000
con.head(20)
```

```python
##11. more_than_6000 컬럼에서 True인 데이터들의 평균, 개수, 편차 등의 정보 출력
# print('평균: ', con[con['more_than_6000'] == True].mean())
# print('개수: ', con[con['more_than_6000'] == True].count())
# print('편차: ', con[con['more_than_6000'] == True].std())

con[con.more_than_6000 == True].describe()
```

```python
##12. more_than_6000 이름의 함수를 생성하고, 6000원이상인 경우 A group, 아니면 B group을 반환하는 함수 생성
# def more_than_6000(x) : 
#     if con['hourly_wage'] >= 6000:
#         print('A group')
#     else :
#         print('B group')
# con['hourly_wage'].apply(lambda x: more_than_6000(x))

def more_than_6000(x: int) -> int :
    if x >= 6000:
        return 'A group'
    else:
        return 'B group'
```

```python
##13. more_than_6000_f 컬럼 생성하고 more_than_6000 함수의 결과를 저장
con['more_than_6000_f'] = con.hourly_wage.map(lambda x: more_than_6000(x))
```

```python
##14. 지금까지의 결과 상위 10개를 출력
con.head(10)
```

```python
##15-1. more_than_6000가 True인 데이터의 지역과 시간당 급여를 가진 새로운 데이터프레임 생성(data2)
##15-2. data2 데이터를 시간당 급여 순으로 정렬 (높은순)
# new = con[con['more_than_6000'] == True]['area1']
# new2 = con[con['more_than_6000'] == True]['hourly_wage']
# data2 = df(index=[new,new2])
# data2

data2 = con[con.more_than_6000 == True][['area1','hourly_wage']]
data2 = data2.sort_values(by='hourly_wage' , ascending= 0)
data2
```

```python
##16. data2를 darta2.csv 파일로 저장
data2.to_csv('data2.csv', index=False)
```

```python
##17. 시간당 급여를 histogram 으로 표시 (matplotlib hist() 사용)
# con['hourly_wage'].hist()

# bins: 너비
con.hourly_wage.hist(bins=10)
plt.pyplot.show()
```

```python
##18. 시간당 급여를 box 차트로 표시 
# boxplot = df.boxplot(con['hourly_wage'])

con.boxplot(column='hourly_wage', return_type='dict')
```

```python
##19. 시간당 급여를 box 차트로 표시(이름순으로)
con.boxplot(column='hourly_wage', by='name')
```

```python
##19. 시간당 급여를 box 차트로 표시(지역순으로)
con.boxplot(column='hourly_wage', by='area1')
```
